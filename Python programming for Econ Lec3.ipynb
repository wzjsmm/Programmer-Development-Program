```python
import numpy as np
import matplotlib.pyplot as plt

T = 200  # the end period
α = 0.9  # like saving rate
x = np.empty(T+1) # an empty Numpy array, to store all x_t
x[0] = 0          # initial value of x
ϵ_values = [] # an empty array

for i in range(T+1):      # create an random iid shock
    e = np.random.randn()
    ϵ_values.append(e)

for t in range(T):
    x[t+1] = α * x[t] + ϵ_values[t+1]

plt.plot(x)
plt.show()
```

```python
α_values = [0, 0.8, 0.98]
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T): 
    α = α_values[0]
    x[t+1] = α * x[t] + np.random.randn()

plt.plot(x,label = "α = 0")

for t in range(T): 
    α = α_values[1]
    x[t+1] = α * x[t] + np.random.randn()
plt.plot(x, label = "α = 0.8")

for t in range(T): 
    α = α_values[2]
    x[t+1] = α * x[t] + np.random.randn()
plt.plot(x, label = "α = 0.98")
plt.legend()
plt.show()
```

```python
α_values = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)

for α in α_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
    plt.plot(x, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```

label 参数用于设置曲线的标签，这个标签会显示在图例中（如果调用了 plt.legend()）。
f'...' 是 Python 的格式化字符串（f-string），可以将变量的值嵌入到字符串中。
$...$ 是 LaTeX 语法的数学公式格式，用于在图例中显示数学符号。
Matplotlib 支持在字符串中嵌入 LaTeX 语法来生成漂亮的数学表达式。
{α} 表示插入变量 α 的值
在 label=f'$\alpha = {α}$' 中，使用 两个"\\" 是因为字符串中嵌入了 LaTeX 表达式，而在 LaTeX 中 \ 是一个转义字符。

```python
α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    if x[t] < 0:
        x[t] = -x[t]
    else:
        x[t] = x[t]
    x[t+1] = α * x[t] + np.random.randn()
    
plt.plot(x)
plt.show()
```

```python
α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    if x[t] < 0:
        abs_x = - x[t]
    else:
        abs_x = x[t]
    x[t+1] = α * abs_x + np.random.randn()

plt.plot(x)
plt.show()
```

#还是新设立一个变量更好，这样子哪怕出错也可以保证原数据x还是存在，所以另设abs_x更好

```python
α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    abs_x = - x[t] if x[t] < 0 else x[t] #这里更简便 可以直接用公式调用if
    x[t+1] = α * abs_x + np.random.randn()

plt.plot(x)
plt.show()
```

```python
# A monte carlo simulation to estimate π
n = 1000000 # sample size for Monte Carlo simulation

count = 0
for i in range(n):

    # drawing random positions on the square
    u, v = np.random.uniform(), np.random.uniform()

    # check whether the point falls within the boundary
    # of the unit circle centred at (0.5,0.5)
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)

    # if it falls within the inscribed circle, 
    # add it to the count
    if d < 0.5:
        count += 1

area_estimate = count / n

print(area_estimate * 4)  # dividing by radius**2
```
